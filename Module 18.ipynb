```python
# Import dependencies
import pandas as pd
```

```python
# Load the iris csv into a dataframe
file_path = "Resources/iris.csv"
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# Drp the class column
new_iris_df = iris_df.drop(['class'], axis = 1)
new_iris_df.head()
```

```python
ordered_iris_df = new_iris_df[['sepal_length', 'petal_length', 'sepal_width', 'petal_width']]
ordered_iris_df.head()
```

```python
# Save new dataframe as csv file 
output_file_path = "Resources/ordered_iris_df.csv"
ordered_iris_df.to_csv(output_file_path, index=False)
```

```python
# Load in shopping_data csv file
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path)
df_shopping.head()
```

```python
# Columns
df_shopping.columns
```

```python
# List dataframe datatypes
df_shopping.dtypes
```

```python
# Find null values
for column in df_shopping.columns:
    print(f"Column{column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
```

```python
# Find duplicate entries
for column in df_shopping.columns:
    print(f"Duplicate entries {df_shopping.duplicated().sum()}")
```

```python
# Transform Genre column
def changeGenre(member):
   if member == "Yes":
       return 1
   else:
       return 0
df_shopping["Card Member"] = df_shopping["Card Member"].apply(changeGenre)
df_shopping.head()
```

```python
# Transfer local columns
df_shopping['Annual Income'] = df_shopping['Annual Income'] / 1000
df_shopping.head()
```

```python
new_df = df_shopping.loc[(df_shopping!=0).any(axis=1)]
new_df
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```
